```python
data_per_year = []

for tutor in tutors:
    
    try:
        name = tutor.find('a', {'class': 'b link_dark2 link'}).get_text() 
        link = 'https://www.hse.ru' + tutor.find('a', {'class': 'b link_dark2 link'}).get('href')
    except:
        name = tutor.find('div', {'class': 'b-greetings__person_data small'}).get_text() 
        link = None
    
    status = tutor.find('div', {'class': 'b-greetings__person_data small'}).find('div').get_text()
    
    data_per_tutor = [name, link, status]
    
    data_per_year.append(data_per_tutor)
    
print(data_per_year[:15])
```
