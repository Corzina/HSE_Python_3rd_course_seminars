**Шаг 11.** Проведите небольшую предобработку: создайте колонку `DateTime` с человеческим отображением времени последнего визита. Затем разделите эту колонку на две: в одной будет дата визита (`Date`), в другой - время (`Time`). Удалите две исходные колонки (самую первую с UNIX-временем и колонку `DateTime`).

```python
Date = pd.to_datetime(data['В сети последний раз'], unit ='s').dt.strftime("%Y.%m.%d")
Date
Time = pd.to_datetime(data['В сети последний раз'], unit ='s').dt.strftime("%H.%M.%S")
Time
data['День'] = Date
data['Время'] = Time
data.drop(['В сети последний раз'], axis='columns', inplace=True)
data
```

```python
data
```

**Шаг 12.** Проведите небольшую предобработку: перекодируйте данные в колонке с полом из 1-2 в женский-мужской. Заментие старую колонку новой.

```python
data['пол'] = pd.to_datetime(df['date'], unit='s').apply(lambda x: str(x).split()[0])
```

```python
def recode(x):
    if x == 1:
        return 'женский'
    if x == 2:
        return 'мужской'
    
yoba = data['Пол'].apply(recode)
data['Пол'] = yoba
data
```

**Шаг 13.** Проведите небольшую предобработку: создайте переменную, которая будет отражать, в Москве человек находится или нет, на основе значений из колонки с городом. В этой колонке должны быть значения "Москва" и "Не Москва", сама колонка пусть называется `Moscow`.

```python
def Muscovite(x):
    if x == 'Москва':
        return 'Москва'
    else:
        return 'не Москва'
    
yoba1 = data['Город'].apply(Muscovite)
data['Moscow'] = yoba1
data
```
