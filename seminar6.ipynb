```python
# создайте новую переменную mathematician, в которой будут находиться сведения о том, 
# что студент в тесте по математике набрал баллов больше, чем в тесте по чтению 
# (1 - если так, 0 - иначе; подсказка: скорее всего, вам понадобится создать ещё одну промежуточную переменную). 
# Какой процент таких студентов находится в этой выборке?

students['highmath'] = students['math score'] - students['reading score']

def recode(highmath):
    if highmath > 0:
        return 1
    elif highmath <= 0:
        return 0

students['mathematician'] = students['highmath'].apply(recode)
students
```
